```python
import torch

# Tensor Indexing

batch_size = 10
features = 25
x = torch.rand((batch_size, features))

print(x[0].shape)
print(x[:, 0].shape)
print(x[2, 0:10].shape)

# Fancy indexing
x = torch.arange(10)
indices = [2, 5, 8]
print(x[indices])

x = torch.rand((3, 5))
rows = torch.tensor([1, 0])
cols = torch.tensor([4, 0])
print(x[rows, cols].shape)
print(x, x[rows, cols])
  
x = torch.arange(10)
print(x[(x < 2) | (x > 8)])
print(x[x.remainder(2) == 0])

# More advanced indexing
print(torch.where(x > 5, x, x*2))
print(torch.tensor([0, 0, 1, 2, 2, 3, 4]).unique())
print(x.ndimension())
print(x.numel())
```

```python
# Tensor Reshaping

x = torch.arange(9)
x_3x3 = x.view(3, 3)
x_3x3 = x.reshape(3, 3)
print(x_3x3)

y = x_3x3.t()
print(y)
print(y.contiguous().view(9))

x1 = torch.rand((2, 5))
x2 = torch.rand((2, 5))
print(torch.cat((x1, x2), dim=0).shape)
print(torch.cat((x1, x2), dim=1).shape)

z = x1.view(-1)
print(z.shape)

batch = 64
x = torch.rand((batch, 2, 5))
z = x.view(batch, -1)
print(z.shape)

z = x.permute(0, 2, 1)
print(z)

print(torch.arange(10).unsqueeze(0).unsqueeze(1))
```
